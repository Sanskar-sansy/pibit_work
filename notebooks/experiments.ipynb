```python
import sys
sys.path.insert(0, '..')

from app.utils.config import load_config
from app.persistence.database import DatabaseManager
import pandas as pd

cfg = load_config('../configs')
db = DatabaseManager(url=cfg.database.url)
experiments = db.list_experiments()
print(f'Found {len(experiments)} experiments')
```

```python
# Load trajectory for latest experiment
if experiments:
    exp = experiments[0]
    iterations = db.get_experiment_iterations(exp.id)
    df = pd.DataFrame([
        {'iter': it.iteration_number, 'f1': it.val_f1, 
         'accepted': it.accepted, 'strategy': it.mutation_strategy}
        for it in iterations
    ])
    display(df)
```
